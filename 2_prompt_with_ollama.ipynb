```python
import re
import os
from PyPDF2 import PdfReader
import string

def extract_text_from_pdf(mypath):
    try:
        pdf_reader = PdfReader(mypath)
        full_text = ""
        text = ""
        for page_num in range(len(pdf_reader.pages)):
            if page_num >= 30 :
                text = pdf_reader.pages[page_num].extract_text()
                text = text.strip()
                print("text length: "+ str(len(str(page_num))))
                print("page num: " + str(page_num))
                text = clean_text(text)

                print(page_num+1)
            #if page_num == 30:
                #full_text += text
            #else:                
            full_text += text
        #print(full_text)
        return(full_text)
    except Exception as e:
        print(f"Error reading {mypath}: {e}")
        return ""

def clean_text(text):
    text = text.lower()
    text = re.sub('bürgerschaft der freien und hansestadt hamburg', '', text)
    text = re.sub('bürgerschaft der freien und hansest adt hamburg', '', text)
    text = re.sub('– 22. wahlperiode', '', text)
    text = re.sub('drucksache 22/'+ pdf_file[3:-4], '', text)
    text = re.sub('\[.*?\]', '', text)
    #text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    #text = re.sub('\w*\d\w*', '', text)
    text = re.sub('-\n\n\n', '', text)
    text = re.sub('-\n\n', '', text)
    text = re.sub('-\n', '', text)
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)   
    text = re.sub('\nder', 'der', text)
    text = re.sub('\nDER', 'DER', text)
    text = text.strip()

    

    return text

#process text file, only top 10 pages
for pdf_file in Pdf_files:
    print("process file number:" + pdf_file)
    pdf_path = os.path.join(mypath, pdf_file)

    full_text = extract_text_from_pdf(pdf_path)
    txt_filename = pdf_file[:-4] + "_7.txt"
    if not full_text.strip() == "":
        with open(os.path.join(mypath + '/txt/', txt_filename), 'w', encoding='utf-8') as txt_file:
                txt_file.write(full_text)
```
